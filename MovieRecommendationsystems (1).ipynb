````python
def give_rec(title, model):

    # Step 1: Get the index of the given movie title
    # 'indices' is a mapping: movie title → row index
    # Example: "Avatar" → 0
    idx = indices[title]


    # Step 2: Get similarity scores of this movie with ALL other movies
    # model[idx] → returns a row from similarity matrix
    # Each value = similarity between this movie and another movie

    # list(...) → converts from array to Python list (easier to work with)

    # enumerate(...) → attaches index to each similarity score
    # Output format: [(movie_index, similarity_score), ...]


    model_scores = list(enumerate(list(model[idx])))


    # Step 3: Sort movies based on similarity score (highest first)
    # key=lambda x: x[1] → sort using similarity score
    # reverse=True → descending order (most similar first)

    model_scores = sorted(model_scores, key=lambda x: x[1], reverse=True)


    # Step 4: Remove the first movie (itself) and take top 10 similar movies
    # model_scores[0] → same movie (similarity = 1.0)
    # So we skip it and take next 10

    model_scores = model_scores[1:11]


    # Step 5: Extract only movie indices from (index, score) tuples
    # We only need indices to fetch movie titles later

    movie_indices = [index[0] for index in model_scores]


    # Step 6: Return movie titles using indices
    # .iloc → used for integer-based indexing
    # We pass movie_indices to get corresponding movie names

    return data['original_title'].iloc[movie_indices]
````

````python
# Function to extract the director's name from the 'crew' column
def get_director(x):

    # 'x' is expected to be a list of dictionaries
    # Example:
    # [
    #   {'job': 'Director', 'name': 'James Cameron'},
    #   {'job': 'Producer', 'name': 'Jon Landau'}
    # ]

    for i in x:
        # Loop through each dictionary in the list
        # Each 'i' represents one crew member

        if i['job'] == 'Director':
            # Check if this crew member's role is "Director"

            return i['name'].strip()
            # Return the director's name
            # .strip() removes extra spaces from beginning/end

    else:
        # This 'else' belongs to the FOR loop (not the if)
        # It executes only if the loop completes WITHOUT finding a 'Director'

        return np.nan
        # Return NaN (missing value)
        # Useful for handling cases where director info is absent
````

````python
def get_list(x):

    names = []
    # Initialize an empty list to store extracted names



    # Check if x is actually a list (safe handling)
    # Because sometimes data may be NaN or not in expected format
    if isinstance(x, list):

        for i in x:
            # Loop through each element in the list
            # Each 'i' is expected to be a dictionary like {'name': 'Action'}

            name = i['name']
            # Extract the 'name' value from the dictionary

            name = name.strip()
            # Remove leading/trailing spaces (e.g., " Action " → "Action")

            name = name.lower()
            # Convert to lowercase for consistency
            # Helps avoid duplicates like "Action" vs "action"

            names.append(name)
            # Add the cleaned name to the list


    # Return only the first 3 names
    # This helps reduce noise and keeps only the most important features
    return names[:3]
````

````python
# Define a function that takes a movie title as input and returns similar movies
def recommend_movies(movie):
    # Calculate the Pearson correlation between the specified movie's ratings
    # and all other movies in the matrix
    # This identifies movies that users rated in a similar way (high or low)
    similarity = moviemat.corrwith(moviemat[movie])

    # Sort the results so the most similar movies (correlation closest to 1.0)
    # appear at the top of the list
    return similarity.sort_values(ascending=False)
````

````python
# ==============================
# IMPORT LIBRARIES
# ==============================

import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
from difflib import get_close_matches


# ==============================
# STEP 1: CREATE USER-ITEM MATRIX
# ==============================

pivot_table = data.pivot_table(
    index='title',      # rows → movies
    columns='userId',   # columns → users
    values='rating'
).fillna(0)


# ==============================
# STEP 2: CONVERT TO SPARSE MATRIX
# ==============================

features_matrix = csr_matrix(pivot_table.values)


# ==============================
# STEP 3: APPLY MATRIX FACTORIZATION (SVD)
# ==============================

svd = TruncatedSVD(n_components=50, random_state=42)

# Convert movies into latent feature space
latent_matrix = svd.fit_transform(features_matrix)


# ==============================
# STEP 4: COMPUTE SIMILARITY
# ==============================

similarity = cosine_similarity(latent_matrix)


# ==============================
# STEP 5: RECOMMENDATION FUNCTION (FUZZY MATCH)
# ==============================

def recommend_movies(movie_name, n_recommendations=10):

    # ------------------------------
    # FIND CLOSEST MATCH (FUZZY SEARCH)
    # ------------------------------

    all_titles = pivot_table.index.tolist()

    # Convert to lowercase for better matching
    all_titles_lower = [title.lower() for title in all_titles]

    matches = get_close_matches(movie_name.lower(), all_titles_lower, n=1, cutoff=0.3)

    if not matches:
        return "❌ No similar movie title found!"

    # Get original title
    matched_movie = all_titles[all_titles_lower.index(matches[0])]

    print(f"✅ Showing results for: {matched_movie}")


    # ------------------------------
    # GET SIMILAR MOVIES
    # ------------------------------

    # Get the index (row number) of the matched movie
    # pivot_table.index contains movie titles
    # get_loc(...) finds the integer position of that movie
    movie_idx = pivot_table.index.get_loc(matched_movie)


    # Get similarity scores of this movie with ALL other movies
    # similarity[movie_idx] → gives a list of similarity values
    # enumerate(...) → converts it into (index, score) pairs
    # Example: [(0, 1.0), (1, 0.32), (2, 0.78), ...]
    sim_scores = list(enumerate(similarity[movie_idx]))


    # Sort movies based on similarity score (highest first)
    # x[1] → similarity value
    # reverse=True → descending order (most similar first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    # Skip the first movie because it is the same movie (similarity = 1.0)
    # Then take top 'n_recommendations' movies
    sim_scores = sim_scores[1:n_recommendations+1]


    # Extract only movie indices (ignore similarity scores)
    # Each element is (index, score) → we take index
    movie_indices = [i[0] for i in sim_scores]


    # Convert indices back to movie titles using pivot_table index
    # .tolist() → converts result to a Python list
    return pivot_table.index[movie_indices].tolist()


# ==============================
# STEP 6: TEST IT
# ==============================

print(recommend_movies("Iron man"))
````
